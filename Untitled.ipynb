```python
combined = pd.merge(puma, wages, how='left', on=['occupation_id', 'occupation_name'])
```

```python
combined.info()
```

```python
combined.head()
```

```python
# Check to see that our uniques are intact
combined.nunique()
```

```python
# No duplicates!!!!
combined.duplicated(subset=None, keep='first').sum()
```

```python
avg_loc_wage = combined.groupby(['puma_name'])['average_wage'].mean()
```

```python
avg_loc_wage = avg_loc_wage.reset_index()
avg_loc_wage
```

```python
highest_paid_per_location = combined.sort_values(by='average_wage',ascending=False).groupby('puma_name').head(1)[['puma_id', 'puma_name','average_wage', 'occupation_name','national_average_wage' ]]
```

```python
highest_paid_per_location
```

```python
#Testing the accuracy of the group
combined[combined['occupation_name'] == 'Music directors and composers'].sort_values(by='average_wage', ascending=False)
```

```python
highest_paid_per_location[highest_paid_per_location['occupation_name'] == 'Surgeons']
```

```python
final_table = pd.merge(avg_loc_wage, highest_paid_per_location, on='puma_name')
```

```python
final_table
```

```python
final_table = final_table.drop('puma_id', axis=1)
```

```python
final_table.rename(columns = {'puma_name': 'location_name(PUMA)', 'average_wage_x':'location_avg_wage','occupation_name' :'highest_wage_occupation(per_location)', 'average_wage_y':'highest_salary(per_location)', 'national_average_wage':'avg_national_wage'}, inplace=True)
```

```python
swap_cols = ['location_name(PUMA)','location_avg_wage','highest_wage_occupation(per_location)', 'highest_salary(per_location)','avg_national_wage']
final_table.reindex(columns=swap_cols)
```
